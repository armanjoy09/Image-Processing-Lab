```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load image
imgg=Image.open("/content/Image.jpg")
plt
```

```python
# check the shape of image
imgg.size
```

```python
# resize the image
resized_imgg=imgg.resize((224,224))
```

```python
# check the resized image size
resized_imgg.size
```

```python
# convert the resized image to array
imgg_array=np.array(resized_imgg)
```
