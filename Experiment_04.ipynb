```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
img=cv2.imread('/content/Image.jpg')
```

```python
# Apply Gaussian blur with a specified kernel size (odd and positive)
ksize=(5,5)
gaussian_blur=cv2.GaussianBlur(img,ksize,0)
```

```python
# Display the original and blurred images side by side
fig,ax=plt.subplots(1,2,figsize=(20,20))
ax[0].imshow(img)
ax[0].set_title('original image')
ax[0].axis('off')

ax[1].imshow(img)
ax[1].set_title('gaussian blurred image')
ax[1].axis('off')
plt.show()
```
