```python
#rreplacing zero values with null values
Replace_0 = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
dataset[Replace_0] = dataset[Replace_0].replace(0, np.nan)
```

```python
dataset.isnull().sum()
```

```python
#filling null values with the mean for each column
columns_fill = ['Glucose', 'BloodPressure', 'SkinThickness','BMI', 'Insulin', 'Age']
mean_values = dataset[columns_fill].mean()
dataset[columns_fill] = dataset[columns_fill].fillna(mean_values)
```
