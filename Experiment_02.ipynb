```python
# if i want plot first 16 images from this folder(for more learning)
import matplotlib.pyplot as plt
import numpy as np

# Get class names
class_names = folder.class_names

# Take one batch from the dataset
for images, labels in folder.take(1):  # First batch
    plt.figure(figsize=(12, 12))  # Big figure

    # Plot first 16 images from this batch
    for i in range(16):
        ax = plt.subplot(4, 4, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))  # Convert to NumPy for plotting
        plt.title(class_names[labels[i]])  # Label
        plt.axis("off")

    plt.tight_layout()
    plt.show()
```
