```python
# First method to load an image using cv2
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('/content/Image.jpg')
plt.imshow(img)
```

```python
# Second method to load an image using PIL
from PIL import Image
img_path='/content/Image.jpg'
img=Image.open(img_path)
img
```

```python
# Third method to load an image
from tensorflow.keras.preprocessing import image
im=image.load_img('/content/Image.jpg',target_size=(224,224))
plt.imshow(im)
```
